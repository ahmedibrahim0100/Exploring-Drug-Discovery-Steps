```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
from ipywidgets import interact
```

```python
@interact
def show_molecule(smiles=''):
  if not smiles:
    return
  mol = Chem.MolFromSmiles(smiles)
  display(mol)
  if not mol:
    return
  mol = Chem.AddHs(mol)
```
